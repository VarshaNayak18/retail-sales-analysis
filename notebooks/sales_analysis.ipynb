```python
# Order Date is a String
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df.dtypes
```

```python
# Create new column for month and year
df["Month"] = df["Order Date"].dt.month
df["Year"] = df["Order Date"].dt.year
```

```python
# To verify the creation of columns
df.head(3)
```

```python
# Sales by category
category_sales = df.groupby("Category")["Sales"].sum()
```

```python
category_sales
```

```python
# Visualization of Sales by Category

category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

* Insight: Technology category contributes the highest total sales compared to Furniture and Office Supplies.
