```python
# Order Date is a String
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
# Create new column for month and year
df["Month"] = df["Order Date"].dt.month
df["Year"] = df["Order Date"].dt.year
```

#### Sales by Category Analysis

```python
# Sales by category
category_sales = df.groupby("Category")["Sales"].sum()
```

```python
category_sales
```

```python
# Visualization of Sales by Category

category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

* Insight: Technology category contributes the highest total sales compared to Furniture and Office Supplies.

#### Profit by Category Analysis

```python
category_profit = df.groupby("Category")["Profit"].sum()
```

```python
category_profit
```

```python
# Visualization of Profit by Category
category_profit.plot(kind="bar")
plt.title("Profit by Category")
plt.ylabel("Total Profit")
plt.xlabel("Category")
plt.show()
```

* Insight: Technology category generates the highest profit among all categories.
Furniture shows relatively lower profitability compared to its sales.

#### Profit by Region Analysis

```python
region_profit = df.groupby("Region")["Profit"].sum()
```

```python
region_profit
```

```python
# Visualization of Profit by Region
region_profit.plot(kind="bar")
plt.title("Profit by Region")
plt.ylabel("Total Profit")
plt.show()
```

* Insight: The West region contributes the highest overall profit.

#### Sub-categories causing loss

```python
loss_products = df.groupby("Sub-Category")["Profit"].sum().sort_values()
```

```python
loss_products
```

```python
# Visualization of products causing losses
loss_products.head(10).plot(kind="bar")
plt.title("Top Loss-Making Sub-Categories")
plt.ylabel("Profit")
plt.show()
```

* Insight: Some sub-categories generate negative profit despite having significant sales.
The reason maybe high discounts, operational costs, or inefficient pricing strategies.

#### Sales vs Profit

```python
#  Scatter Plot
plt.scatter(df["Sales"], df["Profit"])
plt.title("Sales vs Profit Relationship")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

* Insight: Higher sales do not always lead to higher profit. Some transactions with high sales still generate low or negative profit.

#### Dicount vs Profit

```python
plt.scatter(df["Discount"], df["Profit"])
plt.title("Discount vs Profit Relationship")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

* Insight: Higher discounts tend to reduce profit and may even lead to losses. Excess discount strategies may negatively impact business profitability.

```python
# Correlation Heatmap
corr = df[["Sales", "Profit", "Discount", "Quantity"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

* Insight: The correlation heatmap shows relationships between variables such as sales, profit, discount, and quantity. Discounts show a negative relationship with profit, confirming that higher discounts reduce profitability.
